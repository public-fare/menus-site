```python
import re
import json

def parse_dish_links(s):
    dbl_quoted_str = re.sub("\'", '\"', s)
    as_json = json.loads(dbl_quoted_str)
    dish_relation = [item for item in as_json if item['rel'] == 'dishes'][0]
    dish_link = dish_relation['href']
    return dish_link

locations_df['dishes_link'] = locations_df.links.map(parse_dish_links)
```

```python
from urllib import parse

def fix_links(l):
    correct_root = 'http://api.menus.nypl.org/'
    orig_url_parsed = parse.urlparse(l)
    new_path = re.split('/', orig_url_parsed.path, maxsplit=2)[-1]
    fixed_link = parse.urljoin(correct_root, new_path)
    return fixed_link

locations_df['dishes_link'] = locations_df.dishes_link.map(fix_links)
```

```python
request_queue = [i for i in locations_df.dishes_link.unique()]
```

```python
import os
import requests
import time
import logging
```

```python
sess = requests.Session()
sess.params.update({'token' : os.environ['MENUS_API_KEY']})
```

```python
logger = logging.getLogger("dishes")
logger.setLevel(20)
loggerFileHandler = logging.FileHandler("dishes.log")
loggerFileHandler.setLevel(20)
loggerFormatter = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')
loggerFileHandler.setFormatter(loggerFormatter)
logger.addHandler(loggerFileHandler)
```

```python
returned_data = []

counter = 0

while counter < 4752:
    url_in_use = request_queue.pop()
    r = sess.get(url_in_use)
    returned_data.append(r.content)
    logger.info("Processed: {0}. Status: {1}".format(r.request.url, r.status_code))
    time.sleep(0.5)
```
