NB: This data set is not restricted to the same 1851-1921 time period so first, we need to trim it down to only the relevant menus

```python
date_restricted_geolocations = geolocations[geolocations.menu_id.isin(menus_df.index)]
```

```python
date_restricted_geolocations.head()
```

```python
date_restricted_geolocations.info()
```

Let's get rid of columns irrelevant to us …

```python
del date_restricted_geolocations['session_id']
del date_restricted_geolocations['created_at']
del date_restricted_geolocations['updated_at']
del date_restricted_geolocations['comment']
```

If something is noted as "SKIPPED" &mdash; the geolocation action was not completed for some reason? &mdash; then let's not look at those

```python
only_submitted_df = date_restricted_geolocations[date_restricted_geolocations.notes == 'SUBMITTED']
```

```python
only_submitted_df.head()
```

```python
only_submitted_df.info()
```

It looks like there is still data missing &mdash; only 4057 of 6863 rows have lat/lon data. After our conversation I think this is explained by how transit menus are handled. But, because of some inconsistency about how 0 or not a transit type is handled the numbers don't match up right.

As a first pass, I'll take just the submitted data for stationary/non-transit places.

```python
has_lat_data_mask = only_submitted_df.lat.notnull()
```

```python
with_lat = only_submitted_df[has_lat_data_mask]
```

```python
with_lat.info()
```

So, from our set of 12,250 menus, about 4,000 of them have been geocoded through NYPL's app.

```python
with_lat.head()
```

Let's clean up this dataframe a bit before merging it with the rest of the menus info …

```python
del with_lat['notes'] #since all these are 'SUBMITTED' don't need this anymore
```

The ids from the geolocation app are also not particularly relevant so we'll reindex this dataframe by menu id and drop the extra identifier.

```python
with_lat.set_index('menu_id', inplace=True)
```

```python
with_lat.head()
```

```python
menus_1851_to_1921_with_geolocations = pd.merge(menus_df, with_lat, left_index=True, right_index=True)
```

```python
menus_1851_to_1921_with_geolocations.head()
```

Save out this version of the data:

```python
menus_1851_to_1921_with_geolocations.to_csv('../data/menus_1851_to_1921_from_api-geolocated.csv')
```
